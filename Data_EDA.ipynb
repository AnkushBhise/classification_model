```python
## Data Analysis Phase
## MAin aim is to understand more about the data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
## Read the Dataset
dataset=pd.read_csv('Samsung-Dataset.csv')
print(dataset.shape)
dataset.head()
```

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

```python
data = dataset.copy()
(f'Target distibution for features with nan and Without nan value')
for feature in features_with_na:
    data[feature] = np.where(data[feature].isnull(), "NAN", "NOT_NAN")
    Nan_df = data[data[feature] == "NAN"].groupby('target').size()
    NOT_NAN_df = data[data[feature] == "NOT_NAN"].groupby('target').size()
    
    # the same figure for both subplots
    fig = plt.figure(figsize=(4,3),dpi=130)
    fig.suptitle(f'Feature {feature}')

    # plot the first pie chart in ax1
    ax1 = fig.add_subplot(121)
    ax1.pie(Nan_df, labels=Nan_df.index, autopct=lambda p:f'{p:.0f}%, {p*sum(Nan_df)/100 :.0f} count',
            pctdistance=0.5, labeldistance=1.2, explode = (0.05, 0.05))
    ax1.set_title('nan')

    # plot the sencond pie chart in ax2
    ax2 = fig.add_subplot(122)
    ax2.pie(NOT_NAN_df, labels=NOT_NAN_df.index, autopct=lambda p:f'{p:.0f}%, {p*sum(NOT_NAN_df)/100 :.0f} count',
            pctdistance=0.5, labeldistance=1.2, explode = (0.05, 0.05))
    ax2.set_title('Not nan')
    plt.show()    
```

**Observation:** For nan value target is skewed towards 1, as compare in not nan (*Where values are present*) which is bAlance on target 50/50.

***Numerical Variables***

```python
# list of numerical variables
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print(f'Number of numerical variables: {len(numerical_features)} And Toatal Number of variables in dataset {dataset.shape[1]}')

# visualise the numerical variables
dataset[numerical_features].head()
```

**Distibution of continues variable with respect to target**

```python
# recast into long format 
long_dataset = dataset[numerical_features].drop("ID", axis=1).melt(['target'], var_name='feature', value_name='value')
long_dataset.head()
```

```python
# Ploting dist for different features
sns.displot(
    data=long_dataset, 
    x='value', 
    hue='target', 
    kind='kde', 
    fill=True,
    col='feature',
    col_wrap=4, height=1.5
)
```
