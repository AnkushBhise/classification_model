### $Imports$

```python
## Data Analysis Phase
## MAin aim is to understand more about the data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


# show warnings only once
import warnings
warnings.filterwarnings(action='ignore')

## Display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

### $Read-Data$

```python
## Read the Dataset
dataset=pd.read_csv('Samsung-Dataset.csv')
print(dataset.shape)
dataset.head()
```

```python
# Dividing data in variable and target.
# Removing ID column from Variable, as it doesn't give any meanigfull information.
X = pd.DataFrame(dataset.iloc[:, 1:-1])
y = pd.DataFrame(dataset.iloc[:, -1])
print(f"Shape of X is {X.shape} And Shape of y is {y.shape}")
```

### ***Slipt data into train and test with stratification on y***

```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=0, shuffle=True, stratify=y)
X_train.shape, X_test.shape
```

```python
print(f"ytrain unique value count {y_train.value_counts()}")
print(f"ytest unique value count {y_test.value_counts()}")
```

### $Missing-Values$

#### Catagorical Features

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in X_train.columns if X_train[feature].isnull().sum()>1 and X_train[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(X_train[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    dataset[features_nan]=dataset[features_nan].fillna('Missing')
    return dataset
```

```python
dataset=replace_cat_feature(dataset,features_nan)
dataset[features_nan].isnull().sum()
```

#### Numerical Features

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in X_train.columns if X_train[feature].isnull().sum()>1 and X_train[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values
for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(X_train[feature].isnull().mean(),4)))
```

###### Train Data

```python
## Replacing the numerical Missing Values
for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=X_train[feature].median()
    
    ## create a new feature to capture nan values
    X_train[feature+'nan']=np.where(X_train[feature].isnull(),1,0)
    X_train[feature].fillna(median_value,inplace=True)
    
X_train[numerical_with_nan].isnull().sum()
```

```python
X_train.head()
```

###### Test Data

```python
print(f"Test data missing values \n{X_test[numerical_with_nan].isnull().sum()}")
```

```python
for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=X_train[feature].median()
    
    ## create a new feature to capture nan values
    X_test[feature+'nan']=np.where(X_test[feature].isnull(),1,0)
    X_test[feature].fillna(median_value,inplace=True)
    
X_test[numerical_with_nan].isnull().sum()
```

```python
X_test.head()
```

#### $Numerical-Variable-Scaling$
***Scaling all numberical variable between 0 and 1.***

```python
# list of numerical variables
numerical_features = [feature for feature in X_train.columns if X_train[feature].dtypes != 'O']
print(f'Number of numerical variables: {len(numerical_features)} And Toatal Number of variables in dataset {X_train.shape[1]}')

# visualise the numerical variables
X_train[numerical_features].head()
```

```python
scaler=MinMaxScaler(feature_range=(0, 1))
scaler.fit(X_train[numerical_features])
```

```python
X_train[numerical_features] = scaler.transform(X_train[numerical_features])
X_train[numerical_features].head()
```

```python
X_test[numerical_features] = scaler.transform(X_test[numerical_features])
X_test[numerical_features].head()
```

### $Saving-Processed-Data$

```python
X_train.to_csv("data/X_train.csv", index=False)
X_test.to_csv("data/X_test.csv", index=False)
y_train.to_csv("data/y_train.csv", index=False)
y_test.to_csv("data/y_test.csv", index=False)
```
